```python
import numpy as np
import matplotlib.pyplot as plt

def taylor2_method(f, df_dx, y0, x0, xf, h):
    x_values = [x0]
    y_values = [y0]
    
    x = x0
    y = y0
    
    while x < xf:
        y = y + h * f(x, y) + (h**2 / 2) * df_dx(x, y)
        x = x + h
        
        x_values.append(x)
        y_values.append(y)
    
    return x_values, y_values
```

```python
# Inciso b

import math

def f(x, y):
    return 0.0225 * y - 0.0003 * y**2

def df_dx(x, y):
    return 0.0225 - 0.0006 * y
```

```python
y0 = 25
x0 = 0
xf = 120
h = 12

x_taylor2, y_taylor2 = taylor2_method(f, df_dx, y0, x0, xf, h)

plt.plot(x_taylor2, y_taylor2, label='Taylor2')
plt.legend()
plt.xlabel('t')
plt.ylabel('y')
plt.title('Solución numérica de la EDO')
plt.show()

print(f'Población en {xf/h} años: {y_taylor2[-1]}')
```
